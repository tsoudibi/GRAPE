```python
import pandas as pd
df = pd.read_csv('adult.csv')
CAT = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']
NUM = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
LABEL = ['income']
df.head()
```

```python
# apply one-hot encoding catagorical column
from sklearn.preprocessing import OneHotEncoder

new_df = pd.get_dummies(df[CAT]).astype(int)
new_df[NUM] = df[NUM]
new_df[LABEL[0]] = pd.get_dummies(df[LABEL])['income_>50K'].astype(int)

new_df.to_csv('data.txt', index=False, header=False,sep=' ')
# new_df.to_csv('data.csv', index=False, header=False)
new_df
# pd.DataFrame(new_df)
```

```python
pd.DataFrame(range(len(new_df.columns)-1)).to_csv('index_features.txt', index=False, header=False)
pd.DataFrame([len(new_df.columns)-1]).to_csv('index_target.txt', index=False, header=False)
```
